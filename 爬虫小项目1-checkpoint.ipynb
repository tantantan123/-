```python
import requests
import json
import re
import os
from bs4 import BeautifulSoup
```

```python
def get_urllist(html):
    data=re.findall(r'LOLherojs.champion=(.*?);',html)
    heros=json.loads(data[0])['keys']
    namelist=list(heros.values())
    return namlist
```

```python
def get_image(namelist):
    sumlist=[]
    for i in namelist:
        url='https://lol.qq.com/biz/hero/'+i+'.js'
        html=requests.get(url).text
        items=re.findall(r'LOLherojs.champion.'+i+'=(.*?);',html)
        ids=json.loads(items[0])['data']['skins']
        skinlist=[]
        for j in ids:
            skinlist.append(j['id'])
        sumlist.append(skinlist)
    return sumlist        
```

```python
def main():
    url='https://lol.qq.com/biz/hero/champion.js'
    html=requests.get('https://lol.qq.com/biz/hero/champion.js').text
    data=re.findall(r'LOLherojs.champion=(.*?);',html)
    heros=json.loads(data[0])['keys']
    namelist=list(heros.values())
    sumlist=get_image(namelist)
    for i,j in zip(namelist,sumlist):
        root="D://英雄联盟图片//"+i+"//"
        os.mkdir(root)
        for k in j:
            src="https://ossweb-img.qq.com/images/lol/web201310/skin/big"+k+".jpg"
            r=requests.get(src)
            with open(root+k+".jpg",'wb')as f:
                f.write(r.content)
                print('文件保存成功')
```

```python
main()
```

爬取糗事百科的前十页的图

```python
import requests
from bs4 import BeautifulSoup
import time,csv
import re,os
```

```python
def get_url(url):
    headers={
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36'
    }
    try:
        r=requests.get(url,headers=headers)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''
```

```python
def get_imageinfo(html):
    soup=BeautifulSoup(html,'lxml')
    images=[]
    items=soup.findAll('div',{'class':'thumb'})
    for item in items:
        i=item.find('img')
        images.append(i.attrs['src'])
    return images
```

```python
def write_to_file(images):
    root="D://糗事百科//"
    for i,k in zip(range(len(images)),images):
        src="https:"+k
        response=requests.get(src)
        with open(root+str(i)+".jpg",'wb')as fb:
            fb.write(response.content)
            print(f'第{i}张图片保存完成')
```

```python
def main():
    sumlist=[]
    for i in range(10):
        url='https://www.qiushibaike.com/imgrank/page/'+str(i+1)+'/'
        html=get_url(url)
        images=get_imageinfo(html)
        sumlist.extend(images)
    write_to_file(sumlist)
```

```python
if __name__ == '__main__':
    main()
```
